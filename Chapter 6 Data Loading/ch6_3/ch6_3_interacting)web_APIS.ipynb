```python
# 6.3 Interacting with Web APIs
# Example of using the requests library to interact with a web API
import requests
def fetch_data_from_api(url="https://api.github.com/repos/pandas-dev/pandas/issues"):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for bad responses
        return response.json()  # Return the JSON data from the response
    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return None
```

```python
print("Fetching data from the GitHub API...")
resp = fetch_data_from_api()
print(resp)
```

```python
data = resp
# Print the first issue title and body
data[0]['title'] 
```

```python
if data:
    print("First issue title:", data[0]['title'])
    print("First issue body:", data[0]['body'])
# Print the first issue title and body
```

```python
import pandas as pd
# Convert the data to a DataFrame

issues = pd.DataFrame(data, columns=['number', 'title', 
                                     'labels','state' ])
print(issues.head())
```
