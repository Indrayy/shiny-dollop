```python
class my_dialect(csv.Dialect):
    delimiter = ';'  # Use semicolon as the delimiter
    quotechar = '"'
    doublequote = True
    skipinitialspace = True
    lineterminator = '\n'
    quoting = csv.QUOTE_MINIMAL
# Register the custom dialect
csv.register_dialect('my_dialect', my_dialect)
# Read the CSV file with the custom dialect
with open("ex7.csv", "r") as f:
    df_ex7_custom = pd.read_csv(f, dialect='my_dialect', header=None, names=["civil", "science", "military"])
    # Read the CSV file with pandas
df_ex7_custom
# Unregister the custom dialect
```
