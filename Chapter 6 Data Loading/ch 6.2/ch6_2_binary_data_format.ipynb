```python
import pandas as pd
import requests

url = "https://s4.nontonanimeid.boats/anime.csv"

headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    with open("anime.csv", "wb") as file:
        file.write(response.content)

    frame = pd.read_csv("anime.csv")
    print(frame.head())
else:
    print(f"Failed to download: {response.status_code}")
```

```python
import numpy as np
# Create a DataFrame with random data and save it to an HDFStore

frame7 = pd.DataFrame({"yae": np.random.standard_normal(100)})

store = pd.HDFStore("ex5.h5")  # Create an HDFStore object
store["frame7"] = frame7  # Save DataFrame to HDFStore
```

```python
with pd.HDFStore("ex5.h5") as store:
	store["lumine1"] = frame7  # Save another DataFrame to HDFStore
	store["lumine_artifact"] = frame7["yae"]  # Save another DataFrame to HDFStore
```

```python
with pd.HDFStore("ex5.h5") as store:
    store.put("lumine_artifact", frame7["yae"], format="table")  # Save DataFrame to HDFStore in table format
    result = store.select("lumine_artifact", where="index >= 10 and index <=20")
display(result)
# Select rows from the DataFrame stored in HDFStore based on index range
```

```python
with pd.HDFStore("ex5.h5") as store:
    store.put("lumine_artifact", frame7["yae"], format="table")    
    display(store.select("lumine_artifact", 
                     where="index >= 10 and index <=20"))  # Select rows from the DataFrame stored in HDFStore based on index range
```

```python
store.close()  # Close the HDFStore
# prevent the HDFStore from being closed automatically
# prevent leaking file descriptors
```

```python
# shortcut way to open an HDFStore
frame7.to_hdf("ex5.h5", key="frame7", format="table") 
# Save DataFrame to HDF5 file

store.close()  # Close the HDFStore before reading

pd.read_hdf("ex5.h5", key="frame7", 
            where=["index <5 and index <=20"])  
# Load DataFrame from HDF5 file
```
