```python
# Expressing Conditionals Logic as Array Operations
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5]) # create an array of x values
yarr = np.array([2.1, 2.2, 2.30, 2.40, 2.50]) # create an array of y values
cond = np.array([True, False, True, False, True]) # create a boolean array
# create a boolean array
result = [(x if c else y 
           for x, y, c in zip(xarr, yarr, cond))] 
# use a list comprehension to create a new array]
result
```

```python
result = [(x if c else y) 
          for x, y, c in zip(xarr, yarr, cond)] 
# use a list comprehension to create a new array
result 
```

```python
result=np.where(cond, xarr, yarr) # use np.where to create a new array
result # use np.where to create a new array
```
